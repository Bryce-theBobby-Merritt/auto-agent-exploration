Claude: Great question! Let's apply Anthropic's framework to **venture capital sourcing** (finding and evaluating potential investment opportunities):

## VC Sourcing Through the Agent Framework

### 1. **High Complexity** ‚úì
- **Multi-step process**: Scan thousands of companies ‚Üí filter by criteria ‚Üí research founders ‚Üí analyze market ‚Üí assess competitive landscape ‚Üí synthesize investment thesis
- **Diverse data sources**: Crunchbase, LinkedIn, patent databases, news articles, academic papers, GitHub repos, hiring signals, G2/review sites
- **Contextual judgment**: Understanding subtle signals (team quality, market timing, product-market fit indicators) requires connecting disparate information
- **Similar to coding**: Just as going from design doc ‚Üí PR requires orchestrating many steps, going from investment criteria ‚Üí qualified deal flow requires multi-stage research and synthesis

### 2. **High Value** ‚úì‚úì‚úì
- **Enormous upside**: Finding one unicorn can return an entire fund (100-1000x returns)
- **Time arbitrage**: Partners spend 30-50% of time on sourcing; even 20% efficiency gain = massive leverage
- **Competitive advantage**: Better sourcing = seeing deals earlier = better terms and selection
- **Network effects**: Quality sourcing builds reputation ‚Üí more inbound ‚Üí better deals
- **Dollar impact**: Potentially $10M+ in value per hour saved for successful VCs

### 3. **Medium to High Viability** ‚ö†Ô∏è (This is the key question)
**Strengths:**
- LLMs excel at information synthesis, pattern recognition across unstructured data
- Can process far more companies than human analysts (scale advantage)
- Good at initial filtering, competitive analysis, market sizing research
- Strong at writing investment memos once thesis is clear

**Limitations:**
- **Relationship-dependent**: Best deals come through warm intros and network trust (hard to automate)
- **Tacit knowledge**: Recognizing "founder quality" or "market timing" involves gut feel developed over years
- **Signal vs. noise**: Many pattern-matching heuristics ("X for Y" pitches) don't predict success
- **Data recency**: Hottest startups may not be in public databases yet

**Verdict**: Medium-High viability for **research and analysis** phases, Medium viability for **relationship/judgment** phases

### 4. **Low Cost of Error / Low Cost to Prevent Error** ‚ö†Ô∏è‚ö†Ô∏è (Critical concern)
**This is where VC sourcing differs most from coding:**

**High cost of errors:**
- **False negatives** (missing a great company): Could cost $100M+ in returns
- **False positives** (pursuing bad deals): Wastes partner time, opportunity cost on better deals
- **No "unit tests"**: Can't verify if you passed on the next Stripe until years later
- **Irreversible**: Unlike code bugs, you can't patch a missed investment

**Error prevention is expensive:**
- Would still need experienced partners to review agent output
- Hard to validate agent's judgment without doing the work yourself
- Relationship damage if agent-generated outreach is low-quality

**However, some mitigation:**
- Agent could focus on **top-of-funnel** (casting wider net) where error cost is lower
- Use agent for research/memo writing, keep humans for judgment/relationships
- A/B test by comparing agent-sourced deals to traditional methods over time

## Overall Assessment for VC Sourcing Agents

**Best fit**: ‚úì‚úì‚úì for **research and analysis workflows**
- Market mapping and landscape analysis
- Competitive intelligence gathering
- Investment memo drafting
- Portfolio company monitoring

**Medium fit**: ‚úì‚úì for **initial deal screening**
- Scanning larger universe of companies
- Flagging interesting signals (hiring spikes, product launches, founder backgrounds)
- Enriching inbound deal flow with context

**Poor fit**: ‚úó for **relationship building and final judgment**
- Warm introductions and network cultivation
- Final investment decisions
- Founder assessment and due diligence

**The unlock**: Treating it like an **analyst+ tool** rather than a replacement for partner judgment. The value comes from expanding the aperture (seeing 10x more companies) and improving research quality, not from autonomous decision-making.

## Research & Analysis Workflow Breakpoints

### **1. Initial Scope Definition** üö¶ CRITICAL
**Why intervene here:**
- Sets entire trajectory of research
- Bad framing ‚Üí wasted effort on wrong questions
- Investment thesis requires strategic context agents lack

**Human provides:**
- "We're looking for vertical SaaS in healthcare, specifically around provider burnout"
- Portfolio gaps: "We need more enterprise deals to balance consumer bets"
- Market timing intuition: "AI infra is crowded, but AI applications in legal feel early"

**Agent then does:** Comprehensive market scan based on these guardrails

### **2. After Initial Company Filtering** üö¶ MEDIUM-HIGH
**Breakpoint: Agent returns 50-200 companies matching criteria**

**Why intervene here:**
- Prevents wasting time on deep dives into wrong companies

- Catch agent blind spots or misinterpretations

**Human reviews for:**
- "These 5 feel derivative, skip them"

### **3. After Competitive Landscape Mapping** üö¶ MEDIUM
**Breakpoint: Agent produces market map with players positioned**

**Why intervene here:**
- Market maps require strategic interpretation, not just data plotting

**Human provides:**

### **4. Before Investment Memo Drafting** üö¶ CRITICAL

**Why intervene here:**

## Meta-Pattern: When Humans Add Most Value

### **Strategic Breakpoints (CRITICAL):**
1. **Direction setting** - Initial framing and scope
2. **Judgment calls** - Prioritizing conflicting signals
3. **Creative synthesis** - Building investment thesis
4. **Relationship context** - Network effects, reputation considerations

### **Quality Control Breakpoints (MEDIUM):**
1. **Pattern matching** - "I've seen this before" experience
2. **Sanity checks** - Catching agent hallucinations or logic errors
3. **Nuance** - Reading between the lines

### **Efficiency Breakpoints (LOW but useful):**
1. **Course corrections** - Redirecting research mid-stream
2. **Refinement** - Polishing outputs for specific audience

## Workflow Design Principle

**The "Decision Sandwich" Model:**
```
Human: Strategic direction + constraints
  ‚Üì
Agent: Exhaustive research + analysis
  ‚Üì
Human: Judgment + prioritization + creative synthesis
  ‚Üì
Agent: Execution + formatting + supporting detail
  ‚Üì
Human: Final review + relationship activation
```

The key is: **Humans set direction and make non-obvious calls; agents do comprehensive legwork between those decision points.**

The breakpoints should feel like **natural "stop and reflect" moments** where you'd want to check the map before continuing the journey, not arbitrary interruptions.
